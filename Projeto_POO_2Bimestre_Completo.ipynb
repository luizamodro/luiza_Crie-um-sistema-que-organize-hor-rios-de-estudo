## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
from datetime import datetime

class HorarioEstudoluiza:
    def __init__(self, materia, dia, hora_inicio, hora_fim):
        self.materia = materia
        self.dia = dia
        self.__hora_inicio = hora_inicio  # agora √© privado
        self.hora_fim = hora_fim

    # Getter para hora_inicio
    def get_hora_inicio(self):
        return self.__hora_inicio

    # Setter para hora_inicio
    def set_hora_inicio(self, nova_hora):
        self.__hora_inicio = nova_hora

    # L√≥gica interna: calcula dura√ß√£o do estudo
    def duracao_estudo(self):
        formato = "%H:%M"
        inicio = datetime.strptime(self.__hora_inicio, formato)
        fim = datetime.strptime(self.hora_fim, formato)
        duracao = fim - inicio
        return duracao

    def exibir_horario_luiza(self):
        print(f"Mat√©ria: {self.materia}")
        print(f"Dia: {self.dia}")
        print(f"Hor√°rio: {self.__hora_inicio} - {self.hora_fim}")
        print(f"Dura√ß√£o: {self.duracao_estudo()}")
        print("-" * 30)


# Classe derivada (mant√©m o que voc√™ j√° tinha feito antes)
class HorarioEstudoAvancado(HorarioEstudoluiza):
    def __init__(self, materia, dia, hora_inicio, hora_fim, local):
        super().__init__(materia, dia, hora_inicio, hora_fim)
        self.local = local

    def exibir_local(self):
        print(f"Local de estudo: {self.local}")


# Exemplo de uso
estudo1 = HorarioEstudoluiza("Matem√°tica", "Segunda-feira", "08:00", "10:00")
estudo1.exibir_horario_luiza()

# Usando get/set e verificando encapsulamento
print("Hora de in√≠cio original:", estudo1.get_hora_inicio())
estudo1.set_hora_inicio("09:00")
print("Hora de in√≠cio atualizada:", estudo1.get_hora_inicio())
estudo1.exibir_horario_luiza()
```
